```python
import mpld3
mpld3.enable_notebook()
import pandas as pd
import numpy as np
from matplotlib import pyplot
df=pd.read_csv('v2_2.csv', sep=',',header=0)
#df['close']
#df['close'].shift(periods=-1,fill_value=0)

def shift_ticker(ticker):
    ticker_group=df['TickerName'].isin([ticker])
    df.loc[ticker_group,'labels']=df.loc[ticker_group,'close'].shift(periods=-1,fill_value=0)
    
prev_ticker="namesake"
for index,row in df.iterrows():
    ticker=row[25]
    if(ticker!=prev_ticker):
        print(ticker)
        shift_ticker(ticker)
        prev_ticker=ticker
    #else:
     #   print("skipping")
    
#df['labels']
df.head()
```

```python
#categorizing
def category(close,label):
    if((label-close)>=0):
        return 1
    else:
        return 0
    
#vectorizing the operation
diffs_apply = df.apply(
         lambda row: category(row['close'], row['labels']),
         axis=1)

df['category'] = diffs_apply


#checking for skewness
df_train=df.iloc[0:22000]
df_test=df.iloc[22001:]
#df_train.hist(['RS20'])
df.hist(['category'])
pyplot.show()









#cleaning


clean_group=df['labels'].isin([0])
cleaner=df.loc[clean_group,'labels']
print(cleaner.keys())
df=df.drop(cleaner.keys(),axis=0)
labels_df=df['labels']
category_df=df['category']
print(labels_df)
df=df.drop(['Date','TickerName','open','high','low','volume','labels','category'],axis=1)
df.head()
```

```python
#converting to Numpy
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()

data=df.to_numpy(dtype='int64')
data_scaled=min_max_scaler.fit_transform(data)

labels=labels_df.to_numpy(dtype='int64')
categories=category_df.to_numpy(dtype='int64')



#splitting data
X=data_scaled
y=categories
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2, random_state=1)
X_train, X_val, y_train, y_val= train_test_split(X_train, y_train, test_size=0.5, random_state=2)

print(data_scaled)
print(categories)
print(f'labels are{labels}')
```

```python
from tensorflow import keras
from tensorflow.keras import layers
import tensorflow as tf
import numpy as np

inputs=keras.Input(shape=[24,])
embed=layers.Embedding(input_dim=24, output_dim=64)(inputs)
#dense=layers.Dense(32,activation='relu')
#x=dense(inputs)
#x=layers.Dense(256,activation='relu')(x)
#x=layers.Dense(128,activation='relu')(x)
#x=layers.Dense(44,activation='relu')(x)
#x=layers.Dense(50,activation='relu')(x)
#x=layers.Dense(40,activation='relu')(x)
#x=layers.Dense(30,activation='relu')(x)
#x=layers.Dense(24,activation='relu')(x)
x=layers.LSTM(64)(embed)
#x=layers.Dense(30,activation='relu')(x)





outputs = layers.Dense(1,activation='sigmoid')(x)


model = keras.Model(inputs=inputs, outputs=outputs, name="v2_model")

model.summary()
```

```python
from tensorflow.keras.callbacks import Callback

class Histories(Callback):

    def on_train_begin(self,logs={}):
        self.losses = []
        self.accuracies = []

    def on_batch_end(self, batch, logs={}):
        self.losses.append(logs.get('loss'))
        self.accuracies.append(logs.get('acc'))


histories = Histories()
```

```python
import matplotlib.pyplot as plt

model.compile(
    optimizer='adam',
    #keras.optimizers.RMSprop(learning_rate=1),
loss='binary_crossentropy',
metrics=['accuracy']
    
)

history = model.fit(X_train, y_train, batch_size=500, epochs=200, validation_data=(X_val,y_val),callbacks=[histories])

test_scores = model.evaluate(X_test, y_test, verbose=2)
print("Test loss:", test_scores[0])
print("Test accuracy:", test_scores[1])
```

```python
import matplotlib.pyplot as plt
history_dict = history.history
loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']
epochs = range(1, 201)

plt.plot(epochs, loss_values, 'b', label='Training loss')
plt.plot(epochs, val_loss_values, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
acc_values = history_dict['accuracy']
val_acc_values = history_dict['val_accuracy']
plt.plot(epochs, acc_values, 'b', label='Training acc')
plt.plot(epochs, val_acc_values, 'r', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
